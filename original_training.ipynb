Importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
df = pd.read_csv('./mail_data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Check for null values and replace with null string
# df = df.where((pd.notnull(df)), '')
```

```python
df.shape
```

**Label Encoding**

```python
# label spam mail as 0 and ham mail as 1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
labeled_data=le.fit_transform(df['Category'])
df_category = pd.DataFrame(labeled_data, columns=['labeled_category'])
df_category.head()
```

```python
df.drop(['Category'], axis=1 , inplace=True)
```

```python
df=pd.concat([df, df_category], axis=1)
df.head(5)
```

**Separating the data texts and label**

```python
X = df['Message']

Y = df['labeled_category']
```

```python
print(X)
print(Y)
```

# **Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X.shape)
```

### Feature Extraction
_Converting text data into numerical value_

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
print(X_train_features)
```

# Training the Model
Logistic Regression

```python
model = LogisticRegression()
```

```python
## training the logistic regression model with the training data
model.fit(X_train_features, y_train)
```

#### Model Evaluation

```python
# Prediction on training data
y_train_pred = model.predict(X_train_features)
train_acc = accuracy_score(y_true=y_train, y_pred=y_train_pred) *100 # Training Accuracy

# Prediction on testing data
y_test_pred = model.predict(X_test_features)
test_acc = accuracy_score(y_true=y_test, y_pred=y_test_pred) * 100
```

```python
print("Train Accuracy: ", train_acc)
print("Test Accuracy: ", test_acc)
```

# Building a Predictive System

```python
input_mail = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat...", "Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)

num_to_category = lambda x: "Ham" if x==0 else "Spam" if x==1 else "Error"
num_to_category_list = [num_to_category(i) for i in prediction]
print(num_to_category_list)
```
